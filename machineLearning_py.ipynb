```python


#importing libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
x=np.array([1,2,3,4,5]).reshape(-1,1)
y=np.array([50,55,60,80,91])
model=LinearRegression()
model.fit(x,y)
predicted=model.predict(x)
print("Slople",model.coef_[0])
print("intercept:",model.intercept_)
plt.scatter(x,y,color="red")
plt.plot(x,predicted,color="blue")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Linear Regression")
plt.show()
```

```python
#import libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
x=np.array([1,2,3,4,5]).reshape(-1,1)
y=np.array([0,0,0,1,1])
model=LogisticRegression()
model.fit(x,y)
predicted=model.predict(x)
print(predicted)
plt.scatter(x,y,color="red")
plt.plot(x,predicted,color="blue")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Logistic Regression")
plt.show()

```

```python
import numpy as np
from sklearn.linear_model import LinearRegression

# 1. Prepare the data
# Sunlight Hours (x1)
X = np.array([8, 10, 6, 9, 7]).reshape(-1, 1)
# Energy Output (y)
y = np.array([16, 20, 12, 18, 14])

# 2. Calculate Correlation
correlation = np.corrcoef(X.flatten(), y)[0, 1]

# 3. Create and fit the Linear Regression model
model = LinearRegression()
model.fit(X, y)

# 4. Extract coefficients
intercept = model.intercept_
slope = model.coef_[0]

print(f"--- Task 1 Results ---")
print(f"Correlation: {correlation:.2f}")
print(f"Regression Equation: y = {intercept:.2f} + {slope:.2f}x1")

# Quick verification: Predict for 10 hours
prediction = model.predict([[10]])
print(f"Prediction for 10 hours: {prediction[0]:.2f} MW")
```

```python
import numpy as np
from matplotlib import pyplot as plt
from sklearn.tree import DecisionTreeClassifier

x = [
    [25,50000],
    [35,65000],
    [45,80000],
    [55,95000],
    [65,110000]
]

y = [1,1,1,0,0]

model = DecisionTreeClassifier()
model.fit(x,y)

predicted = model.predict([[30,55000]])
print(predicted)

# Taking only age values
ages = [row[0] for row in x]

plt.scatter(ages, y, color="red")

plt.xlabel("Age")
plt.ylabel("Target")
plt.title("Decision Tree")

plt.show()





```

```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

x = np.array([1,2,3,4,5]).reshape(-1,1)
y = np.array([50,55,60,80,91])

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)
model = LinearRegression()
model.fit(x_train, y_train)
predicted = model.predict(x_test)
print(predicted)
```

```python
#MAE(Mean Absoulte Error)
from sklearn.metrics import mean_squared_error
y_true=[3,-0.5,2,7]
y_pred=[2.5,0.0,2,8]
mse_value=mean_squared_error(y_true,y_pred)
print("Mean Squared Error:",mse_value)
```

```python
#R2(R squared)  ssres is the sum of squared errors
#sstot is the Total variation in Actual vales
from sklearn.metrics import r2_score
y_true=[3,-0.5,2,7]
y_pred=[2.5,0.0,2,8]
r_value=r2_score(y_true,y_pred)
print("R2 Score:",r_value)

```
