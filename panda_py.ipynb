```python
import pandas as pd
mydataset={
    "cars": ["BMW","Volve","AuDi"],
    "passings":[3,7,2]

}
myvar=pd.Series(mydataset)
print(myvar)

a=[1,7,2]
myvar=pd.Series(a)
print(myvar[0])

#label in padas
import pandas as pd
a=[1,7,2]
myvar=pd.Series(a,index=["x","y","z"])

print(myvar["y"])

#Using Index in pandas
calories={"day1":420,"day2":385,"day3":500}
myvar=pd.Series(calories,index=["day1","day2"])

print(myvar)

data={
    "calories":[420,550,700],
    "duration":[50,40,4]

}
myvar=pd.DataFrame(data)
print(myvar)
```
